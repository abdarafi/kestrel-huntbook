### 0. What are the inspection commands?

Kestrel has two command to inspect entities in a variable:
- show details of a variable with [INFO](https://kestrel.readthedocs.io/en/latest/language.html#info)
- display entities in a variable with [DISP](https://kestrel.readthedocs.io/en/latest/language.html#disp)

### 1. How to show information like attributes of a variable?

Once you have a variable `varx`, just `INFO varx` to show all details such as attributes of the entities:

```kestrel
# get scheduler process
scheduler = GET process
            FROM file:///tmp/lab101.json
            WHERE [process:pid = 2224]
            START t'2021-04-03T00:00:00Z' STOP t'2021-04-04T00:00:00Z'
          
# show details of variable `scheduler`
INFO scheduler
```

### 2. How to display entities with complete information in a variable?

Just do `DISP varx` to display all entities with all attributes in variable `varx`:

```kestrel
DISP scheduler
```

### 3. How to display select attributes of entities in a variable?

You may find the table above actually displays the same entity (process) multiple times, because the entity is referred in different [records/logs](https://kestrel.readthedocs.io/en/latest/language.html#record), each of which gives different attributes of the entity.

You may not need to know the `id` attribute of the entity, which is a Kestrel internal data structure when caching the entitiy. To select attibutes to display, and to automatically deduplicate entities which share the same attributes, you can use the `ATTR` clause of [DISP](https://kestrel.readthedocs.io/en/latest/language.html#disp):

```kestrel
DISP scheduler ATTR name, pid, command_line
```

### 4. What are the common attributes for a type of entitiy?

You can find the attributes of a type of entitiy in [STIX SCO docs](https://docs.oasis-open.org/cti/stix/v2.1/csprd01/stix-v2.1-csprd01.html#_Toc16070680).

A quickly way is to use `INFO` and get familar with the common attributes:
- `process`: `name`, `pid`, `command_line`
- `file`: `name`, `parent_directory_ref.path`
- `ipv4-addr`: `value`
- `user-account`: `user_id`
- `windows-registry-key`: `key`, `values`
- `network-traffic`: `src_ref.value`, `src_port`, `dst_ref.value`, `dst_port`
