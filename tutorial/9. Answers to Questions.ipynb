```kestrel
proclist = NEW process [ {"name": "cmd.exe", "pid": "123"}
                       , {"name": "explorer.exe", "pid": "99"}
                       , {"name": "firefox.exe", "pid": "201"}
                       , {"name": "chrome.exe", "pid": "205"}
                       ]

browsers = GET process FROM proclist WHERE name IN ('firefox.exe', 'chrome.exe')

DISP browsers ATTR name, pid
```

### 3. Find Connected Entities.ipynb
Exercise: find processes which connected to `13.86.101.172`

```kestrel
ips = GET ipv4-addr
      FROM file:///tmp/lab101.json
      WHERE value = '13.86.101.172'
      START 2021-04-03T00:00:00Z STOP 2021-04-04T00:00:00Z

conns = FIND network-traffic ACCEPTED BY ips

procs = FIND process CREATED conns
```

### 4. Group Entities in a Variable.ipynb
Exercise: group processes by binary/image and count PIDs

```kestrel
svchost = GET process
          FROM file:///tmp/lab101.json
          WHERE name = 'svchost.exe'
          START 2021-04-03T00:00:00Z STOP 2021-04-03T02:00:00Z
          
scheduler = GET process
            FROM svchost
            WHERE command_line = 'C:\Windows\system32\svchost.exe -k netsvcs -p -s Schedule'
            
scheduled_tasks = FIND process CREATED BY scheduler

task_grps = GROUP scheduled_tasks BY binary_ref.name WITH COUNT(pid) AS number_of_procs

DISP task_grps
```

### 5. Apply a Kestrel Analytics.ipynb
Exercise: group and plot

```kestrel
procs = GET process
        FROM file:///tmp/lab101.json
        WHERE parent_ref.name = 'svchost.exe'
        START 2021-04-03T00:00:00Z STOP 2021-04-03T02:00:00Z
        
procs_grps = GROUP procs BY binary_ref.name WITH COUNT(pid) AS number_of_procs

# need to single quote binary_ref.name so Kestrel will not treat binary_ref as a variable
APPLY python://attribute-plot ON procs_grps WITH XPARAM='binary_ref.name', YPARAM=number_of_procs
```
